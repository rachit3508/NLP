### Get all the punctuation from the string.

```python
import string
punc = string.punctuation
print(punc)
```

#### Write a function - to clean the data (remove stop words , remove punctuation) - print clean data

```python
def cleaner(data):
    data_tokens = nltk.word_tokenize(data)
    stop_words = stopwords.words('english')
    punc = string.punctuation
    punc = list(punc)
    for word in data_tokens:
        if ((word in stop_words) | ((word in punc) & (word!='.'))):
            data_tokens.remove(word)
    for i in data_tokens:
        if ((i in punc) & (i!=".")):
            data_tokens.remove(i)
    return data_tokens
```

```python
cleaned_data = cleaner(data)
```

```python
cleaned_data = " ".join(cleaned_data)
```

### Stemming

```python
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")

print("Porter Stemmer")
print(porter.stem('run'))
print(porter.stem('ran'))
print(porter.stem('running'))

print("----------------------------------------------------")

print("Lancaster Stemmer")
print(lancaster.stem('run'))
print(lancaster.stem('ran'))
print(lancaster.stem('running'))

print("----------------------------------------------------")

print("Sbowball Stemmer")
print(snowball.stem('run'))
print(snowball.stem('ran'))
print(snowball.stem('running'))

print("----------------------------------------------------")
```
