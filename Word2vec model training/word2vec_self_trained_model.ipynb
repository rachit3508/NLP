```python
import nltk
f = open('/content/drive/MyDrive/Indian Rap.txt',"r")
file_data = f.read()
```

```python
print(file_data)
```

```python
file_data = file_data.lower()
```

```python
file_data
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
def rem_stop_words(x):
    li=[]
    words = nltk.word_tokenize(x)
    for word in words:
        if(word in stop_words):
            pass
        else:
            li.append(word)
    x = " ".join(li)
    return x
```

```python
file_data = rem_stop_words(file_data)
```

```python
file_data
```

```python
nltk.download('punkt')
sent_tokens = nltk.sent_tokenize(file_data)
```

```python
from gensim.utils import simple_preprocess
list_of_lists = []
for sent in sent_tokens:
  list_of_lists.append(simple_preprocess(sent))
```

```python
list_of_lists
```

```python
import gensim
```

```python
from gensim.models import Word2Vec
model = Word2Vec(window=10,min_count=2)
```

```python
model.train(list_of_lists,total_examples=len(sent_tokens),epochs=20)
```

```python
model.most_similar('raftaar')
```
